8. Вычислите cos(x) и sin(x) на интервале [0,1] с шагом $0.05$, а затем объедините оба массива чисел как строки в один массив.

```python
import numpy as np
mas = np.arange(0,1,0.05)
mascos = np.array([])
massin = np.array([])
print(len(mas))
for i in range (len(mas)):
    mascos = np.append(mascos,np.cos(mas[i]))

for i in range (len(mas)):
    massin = np.append(massin,np.sin(mas[i]))

mas = np.append(mascos,massin)
print(mascos)
print("Значения косинуса\n========================")
print(massin)
print("Значения синуса\n========================")
print(mas)
print("Объединённый массив\n========================")
```
